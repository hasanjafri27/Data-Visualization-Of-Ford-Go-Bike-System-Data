```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import datetime
%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv('clean_file.csv')
```

## Customers vs. Subscribers : Monthly Bike Usage
This pointplot determines the number of trips performed by random customers and subscribers through out every month. The number of trips performed by the random strangers stays somewhat the same, while the number of trips performed by the subscribers are more in the summer months as compared to the winter months.

```python
df_user_month = df.groupby(['start_time_month_name', 'user_type']).size().reset_index();
graph = sb.pointplot(data = df_user_month, x = 'start_time_month_name', y = 0, hue = 'user_type', height = 8.27/5, aspect = (14.70/12)/(8.27/5));
graph.set_title('Customers vs. Subscribers : Monthly Bike Usage', fontsize = 16, fontweight = 'bold');
plt.xlabel('Month', fontsize = 12);
plt.ylabel('Number Of Trips', fontsize = 12);
plt.xticks(rotation = 45);
plt.grid();
```

## Age Per User Type and Gender :
The box plox of age per user type and gender depicts that the 'others' have the highest median age values among both of the user types. Males have the highest interquartile ranges in both user types suggesting that they are the major users. Females in both the user types have the most wide spread whisker values indicating the highest ages in both user types.

```python
graph = sb.boxplot(data = df, x = "user_type", y = 'member_age', hue = "member_gender");
graph.set_title('Age Per User Type and Gender', fontsize = 16, fontweight = 'bold');
graph.set_xlabel('User Type');
graph.set_ylabel('Age');
plt.figure(figsize=[14.70/2, 8.27/6]);
```

## Trip Duration Per User Type and Gender :
This multivariate bar plot shows that the female customers have been on the longest trips as compared to the duration of trips by male customes. This graph proves that the customer trips are longer than the subscriber trips suggesting that subsribers are mostly commuters while random customers are tourists.

```python
graph = sb.barplot(data = df, x = 'duration_min', y = "user_type", hue = 'member_gender');
graph.set_title('Trip Duration Per User Type and Gender', fontsize = 16, fontweight = 'bold');
plt.xlabel('Trip Duration', fontsize = 12);
plt.ylabel('User Type', fontsize = 12);
plt.figure(figsize=[14.70/6, 8.27/2]);
```

## Summary :
There are two types of clients using the system: Subscribers and Customers. Subscribers are primarily daily commuters, having short trips to and from work, who rent a bike on weekdays at 8-9am and 5-6pm. Customers are usually tourists or occassional riders who use the system mainly on weekends. Age is also a factor within user type. Subscribers who fall in the age group between 26-35 years old are the most common age group to use the bike sharing system. The 26-35 years old also lead the spike which occurs across all age groups in October. Subscribers who fall in the 36-45 year old age group are the next most common age group to use the bike sharing system, and follow a similar trend at the 26-35 year olds.
